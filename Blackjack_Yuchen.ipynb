```python
def draw_card(player_deck,dealer_deck,deck):
    '''
    Draw card function for given number of player
    Input:
        deck: usually the default_deck
            Type: dict
        player_deck: deck contains all players cards
            Type: dict
        num: number of players
            Type: int
    Return:
        player_deck: deck contains all players cards
            Type: dict
        dealer_deck: deck contains all dealer cards
            Type: dict
        deck: deck
            Type: dict
    '''
    assert isinstance(player_deck,dict) and isinstance(dealer_deck,dict) and isinstance(deck,dict),'Input type should be dict'
    
    for player,value in player_deck.items():
            kind1 = random.choice(list(deck.keys()))
            value1 = random.choice(deck[kind1])
            card1 = [kind1,value1]
            for kd,val in deck.items():
                if kd == kind1:
                    i = 0
                    for num in deck[kind1]:
                        if num != value1:
                            i+=1
                        elif num == value1:
                            del deck[kind1][i]             
            player_deck[player].append(card1)
    if dealer_deck == None:
        pass 
    else:
        for dealer, value in dealer_deck.items():
            kind2 = random.choice(list(deck.keys()))
            value2 = random.choice(deck[kind2])
            card2 = [kind2,value2]
            for kd,val in deck.items():
                if kd == kind2:
                    i = 0
                    for num in deck[kind2]:
                        if num != value2:
                            i+=1
                        elif num == value2:
                            del deck[kind2][i]  
        dealer_deck[dealer].append(card2)
    
    return player_deck,dealer_deck,deck   
```

```python
def add_cards(player_deck,dealer_deck):
    '''
    Adds up card value for player and dealer
    Input:
        player_deck: deck contains all players cards
            Type: dict
        dealer_deck: deck contains all dealer cards
            Type: dict
    Return:
        sum_player: the sum of player cards
            Type: int
        sum_dealer: the sum of dealer cards
            Type: int
    '''
    assert isinstance(player_deck,dict) and isinstance(dealer_deck,dict),'the input type should be (dict,dict)'
    
    sum_dealer = 0
    sum_player = 0
    print(dealer_deck)
    if dealer_deck != None:
        for name, value in dealer_deck.items():
            for i in value:
                for j in i:
                    if isinstance(j,int):
                        sum_dealer += j 
                    elif (j == 'J'or j == 'Q'or j == 'K'):
                        sum_dealer += 10
                    elif (j == 'A'):
                        if (sum_dealer+10>21):
                            sum_dealer += 1
                        else:
                            sum_dealer += 11
    if player_deck != None:
        for name, value in player_deck.items():
            for i in value:
                for j in i:
                    if isinstance(j,int):
                        sum_player += j 
                    elif (j == 'J'or j == 'Q'or j == 'K'):
                        sum_player += 10
                    elif (j == 'A'):
                        if (sum_player+10>21):
                            sum_player += 1
                        else:
                            sum_player += 11
        
    return int(sum_player),int(sum_dealer)
```
